```python
import csv

#실습 1
def seoul():
    f=open('seoul.csv' , 'r' , encoding='cp949')
    data = csv.reader(f)
    header = next(data)
    max_temp = -999
    min_temp = -999
    max_diff = 0
    list = [["",0],["",0],["",0]]   #일교차 날짜 , 값 , 최저기온 날짜, 값 , 최고기온 날짜, 값

    for row in data:
        min_temp = float(row[-2])
        if (list[1][1]>min_temp):
            list[1][1] = min_temp
            list[1][0] = row[0]
        max_temp = float(row[-1])
        if (list[2][1]<max_temp):
            list[2][1] = max_temp
            list[2][0] = row[0]
        max_diff = max_temp-min_temp
        if (list[0][1]<max_diff):
            list[0][1] = max_diff
            list[0][0] = row[0]

    print("일교차가 가장 큰 날\n" + list[0][0])
    print("최저기온이 가장 낮은 날\n" + list[1][0])
    print("최고기온이 가장 높은 날\n" + list[2][0])

#실습 2
def subway():
    f = open('subway.csv', 'r', encoding='cp949')
    data = csv.reader(f)
    next(data)
    line = "1호선"   #1호선 부터 시작
    list = [["", 0], ["", 0], ["", 0]]  # (역명 , 이용객수)
    list_min = [["",999999999], ["", 999999999], ["", 99999999]]  # (역명 , 이용객수)
    for row in data:
        p_number = int(row[4].replace("," ,""))+int(row[5].replace("," ,"")) #승객수
        if(line == row[1]): #호선이 같으면
            if p_number > list[0][1]:   #1순위
                list[2][0] = list[1][0] #2순위 -> 3순위
                list[2][1] = list[1][1]
                list[1][0] = list[0][0] #1순위 -> 2순위
                list[1][1] = list[0][1]
                list[0][0] = row[3]     #1순위에 값 삽입
                list[0][1] = p_number
            elif p_number > list[1][1]: #2순위
                list[2][0] = list[1][0]  # 2순위 -> 3순위
                list[2][1] = list[1][1]
                list[1][1] = p_number   # 3순위에 값 삽입
                list[1][0] = row[3]
            elif p_number > list[2][1]: #3순위
                list[2][1] = p_number
                list[2][0] = row[3]
            #-------가장 많은 역---------#
            if p_number < list_min[0][1]:
                list_min[2][0] = list_min[1][0]  # 2순위 -> 3순위
                list_min[2][1] = list_min[1][1]
                list_min[1][0] = list_min[0][0]  # 1순위 -> 2순위
                list_min[1][1] = list_min[0][1]
                list_min[0][0] = row[3]  # 1순위에 값 삽입
                list_min[0][1] = p_number
            elif p_number < list_min[1][1]:  # 2순위
                list_min[2][0] = list_min[1][0]  # 2순위 -> 3순위
                list_min[2][1] = list_min[1][1]
                list_min[1][1] = p_number  # 3순위에 값 삽입
                list_min[1][0] = row[3]
            elif p_number < list_min[2][1]:  # 3순위
                list_min[2][1] = p_number
                list_min[2][0] = row[3]
        else: #호선 바꼈을 시 list 초기화 , line값 바꿈
            print("이용객이 가장 많은 역 순위(" + line+")")
            print("1." + list[0][0] + "(" + str(list[0][1]) + "명)")
            print("2." + list[1][0]+ "(" + str(list[1][1]) + "명)")
            print("3." + list[2][0]+ "(" + str(list[2][1]) + "명)")
            print("이용객이 가장 적은 역 순위(" + line + ")")
            print("1." + list_min[0][0]+ "(" + str(list_min[0][1]) + "명)")
            print("2." + list_min[1][0]+ "(" + str(list_min[1][1]) + "명)")
            print("3." + list_min[2][0]+ "(" + str(list_min[2][1]) + "명)")
            line = row[1]
            list = [["", 0], ["", 0], ["", 0]]
            list_min = [["",999999999], ["", 999999999], ["", 999999999]]  # (역명 , 이용객수)


print("#실습 1")
seoul()
print("\n" + "#실습 2")
subway()
```
